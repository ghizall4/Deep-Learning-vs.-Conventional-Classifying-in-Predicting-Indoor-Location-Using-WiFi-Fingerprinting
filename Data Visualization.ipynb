```python
#Convert the floor values into numbers
for i in range(len(df[df.columns[523]])):
    if df.iloc[i, 523] == 'Floor_0': df.iloc[i, 523] = round(1)
    elif df.iloc[i, 523] == 'Floor_1': df.iloc[i, 523] = round(2)
    elif df.iloc[i, 523] == 'Floor_2': df.iloc[i, 523] = round(3)
    elif df.iloc[i, 523] == 'Floor_3': df.iloc[i, 523] = round(4)
    else: df.iloc[i, 523] = int(5)
```

```python
df.plot(kind="scatter", x="LONGITUDE", y="LATITUDE", alpha=0.2, figsize=(15,10),
    c="BUILDINGID", colorbar=False,
    sharex=False)
```
